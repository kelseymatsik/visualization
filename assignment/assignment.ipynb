```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./visualization/assignment/data/college_completion.csv")
df.shape
df.head()
```

The data has 3798 rows and 63 columns. The variables contain various information about the colleges in the data set and how long students take to complete their degree.

```python
pd.crosstab(df["control"], df["level"])
```

```python
pd.crosstab(df["control"], df["level"], normalize="all")
```

In the context of the "level" and "control" variables, the greatest number of observations fall under 4-year, private non-profit universities. The least number of observations fall under the 2-year, private non-profit universities. There are also a lot of observations thar represent private for-profit and public universities. Out of the private for-profit schools, the observations are almost evenly split between 2 and 4-year institutions. Out of the public schools, there are several more observations that represent 2-year universities than 4-year ones.

```python
sns.histplot(df["grad_100_value"])
```

```python
sns.boxplot(df["grad_100_value"])
```

```python
sns.kdeplot(df["grad_100_value"])
```

```python
df["grad_100_value"].describe()
```

```python
sns.kdeplot(data=df, x="grad_100_value", hue="control")
```

```python
sns.kdeplot(data=df, x="grad_100_value", hue="level")
```

```python
df.loc[:,["grad_100_value", "level", "control"]].groupby(["control", "level"]).describe()
```

Based on the grouped summary statistics, the following categories of universities have the best graduation rates: private for-profit, 2-year; private non-profit, 2-year; and private non-profit, 4-year universities; based on looking at the mean and first and third quartiles. The kernel density plot supports these conclusions.

```python
df["levelXcontrol"] = df["level"] + df["control"]
sns.kdeplot(data=df, x="grad_100_value", hue="levelXcontrol")
```

Looking at this kernel density plot, 2-year private for-profit universities apprear to have the best graduation rates, as the highest percentage of these universities show that higher percentages of students finish their degree within or before 100% of expected time.

```python
sns.kdeplot(data=df, x="aid_value", hue="level")
```

Looking at the above kernel density plot of aid value grouped by 2-year and 4-year universities, the 2-year university aid value peaks at about half the amount that the 4-year university aid value curve does. This makes sense, as 2-year universities would provide half the amount of aid as 4-year universities because the education program is half the length.

```python
sns.kdeplot(data=df, x="aid_value", hue="control")
```

Looking at the above kernel density plot of aid value grouped by type of university, the public and private for-profit density curves peak at a much lower value compared to that of private non-profit universities. The private non-profit universities curve also appears to have much more variation, meaning the amount of aid that private non-profit universities give to undergraduate students is highly variable compared to that of public and private for-profit universities.

```python
df.loc[:,["aid_value", "level", "control"]].groupby(["control", "level"]).describe()
```
